```python
#Author Alois Mbutura
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt

#reflective surface displacement function
t=np.linspace(0,1e-2,1000)
vibrational_freq=200 #Hz
vibrational_amplitude=10*np.pi 
#vibrational_amplitude-Max peak amplitude from rest for simple harmonic motion expressed in radians as we are
#working with sinusoids.

#initial displacement can even be random as per our frame of reference.
init_disp=3.9e6 #radian

#The init_disp can be seen as a DC/constant value added to the displacement fuction.

phi_o=init_disp+(vibrational_amplitude*np.sin(2*np.pi*vibrational_freq*t))
plt.plot(t,phi_o)
plt.show()
```

```python
#leh-linewidth enhancement factor
leh=3
feedback_factor=5


#Phase equation
#Original equation-light_phase=phi_o-(feedback_factor*np.sin(light_phase+np.arctan2(leh)))
#Rearranging
def f(x):
    val=x + 2*np.cos(x)
    return val

solution = opt.fsolve(f,0.1)
print(solution)
#plt.plot(t,light_phase)
#plt.show()
```
